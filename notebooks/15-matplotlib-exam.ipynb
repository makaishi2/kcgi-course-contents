## Matplotlib課題

下記のシグモイド関数グラフ描画のコードを参考にして、sin関数のグラフを描画して下さい。
但し、xの値域は$[0, 6\pi]$ とします。

```python
# 必要ライブラリimport
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

x = np.linspace(-5, 5, 101)
y = sigmoid(x)

plt.plot(x, y, label='sigmoid')
plt.grid()
plt.legend()
plt.show()
```
